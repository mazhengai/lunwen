```python
import jieba
def chinese_word_cut(mytext):
    return " ".join(jieba.cut(mytext))
```

```python
import pandas as pd
```

```python
df = pd.read_csv('wenben.csv')
```

```python
X = df[['xinxi']]
```

```python
X['cutted_xinxi'] = X.xinxi.apply(chinese_word_cut)
```

```python
fl = open('data3.txt', 'w', encoding='utf8')
for i in X.index:
    a = list()
    print(i)
    for j in X['cutted_xinxi'][i].strip('{}').split(','):
        a.append(j)
    fl.write(str(a))
    fl.write('\n')
fl.close()
```
